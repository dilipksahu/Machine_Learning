```python
# linear with cement but hetroscadastic
# Milti-colinearity between superplasticizer vs flyash and water vs age
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)

r2 = r2_score(y_test,y_pred)

print("mse: {},\nrmse: {},\nr2: {}".format(mse,rmse,r2))
```

```python
residuals = y_test - y_pred
```

```python
sns.distplot(residuals)
```

```python
skew(residuals)
```

```python
plt.figure()
plt.scatter(y_pred, residuals)
plt.show()
```

```python
# Handling Skewness
```

```python
for col in df:
    
    if skew(df[col]) >= 0.1 or skew(df[col]) <= -0.1 and col != "csMPa":
        df[col] = np.sqrt(df[col])
     
    plt.figure()
    sns.distplot(df[col])
    plt.show()
    
    print(col,"-",skew(df[col]))
    print("-----------------------------------------------")
```

```python
df.drop("water",axis=1,inplace=True)
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.3)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
print("r2",r2_score(y_test,y_pred))
residuals = y_test - y_pred
plt.hist(residuals)
```
