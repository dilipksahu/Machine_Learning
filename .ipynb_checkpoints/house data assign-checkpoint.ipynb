# 1) Scatter Plot

```python
df.corr()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x = df["sqft_living"], y = df["sqft_lot"])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x = df["price"], y = df["sqft_lot"])
plt.show()
```

# 2) Histogram

```python
plt.figure(figsize=(7,7))
plt.hist(df["grade"], bins=10)
plt.xlabel("Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
# histogram using pandas
df["sqft_living"].hist(bins=10)
plt.xlabel("Sqft Living")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_sqft_liv = df["sqft_living"].mean()
median_sqft_liv = df["sqft_living"].median()
print(mean_sqft_liv, median_sqft_liv)
```

```python
plt.figure()
plt.hist(df["sqft_living"])
plt.axvline(mean_sqft_liv, color = "red")
plt.axvline(median_sqft_liv, color = "yellow")
plt.show()
```

# 3) Bargraph

```python
df["grade"].value_counts().plot(kind="bar")
plt.xlabel("Grade")
plt.show()
```

# 4) Piechart

```python
df["grade"].value_counts().plot(kind="pie")
plt.show()
```

```python
df["grade"].value_counts().plot(kind="pie" ,autopct="%1.1f%%")
plt.show()
```

```python
df["grade"].value_counts().plot(kind="pie" ,autopct="%1.1f%%" ,explode=(0,0,0,0,0.5,0,0,0,0,0,0,0))
plt.show()
```
