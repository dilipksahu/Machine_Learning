```python
# List
heros = [("Batman",80),("Superman",70),("Spider",90),("Ironman",85)]
```

```python
heros
```

```python
# List -> DataFrame
pd.DataFrame(heros,columns=["Name","Power"])
```

```python
# Dictionary
dic = {"Name":["Dilip","Akshay","Rajesh","Surbhi"],"Math":[80,70,85,90],"science":[75,76,80,95]}
```

```python
dic
```

```python
# Dictionary -> DataFrame
df = pd.DataFrame(dic)
```
