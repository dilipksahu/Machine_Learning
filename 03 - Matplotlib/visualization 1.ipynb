## Fig-size,labels and title

```python
area = [600,800,1000,1200,1500]
price = [15,18,25,35,40]
```

```python
# title,xlabel,ylabel, figsize(width,height)

plt.figure(figsize=(7,6))

plt.plot(area,price)
plt.title("Area vs Price line plot")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(7,6))

plt.plot(area,price)
plt.title("Area vs Price line plot")
plt.xlabel("Area ")
plt.ylabel("Price")

#xlim and ylim
# dbl(()) - inside function passes tuple
plt.xlim((400,1000))
plt.ylim((5,50))
plt.show()
```

# xticks and yticks- to rotate labels

```python
plt.figure(figsize=(7,6))

plt.plot(area,price)
plt.title("Area vs Price line plot")
plt.xlabel("Area")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.yticks(rotation=90)
plt.show()
```

## Multiple Plots

```python
x = np.arange(1,20)

y1 = np.sqrt(x)
y2 = np.log(x)
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(labels=["sqrt","log"])
plt.show()
```

## Histogram plot using Pandas

```python
# pandas = numpy + matplotlib
# bin= no. of bars
cars["width"].hist(bins=10)
plt.xlabel("width")
plt.ylabel("frequency")
plt.show()
```

```python
mean_width = cars["width"].mean()
median_width = cars["width"].median()

print(mean_width,median_width)
```

```python
plt.figure()
plt.hist(cars["width"])
# axvline - plot mean and median
plt.axvline(mean_width,color="yellow")
plt.axvline(median_width,color="red")
plt.show()
```

# Visualization 2 (Seaborn)

```python
import pandas as pd
import numpy as np

# Visualization purpose
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("cars.csv")
correlation = cars.corr()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(correlation , annot=True, cmap="viridis")
plt.show()
```

# Box Plot

## To plot the Outliers

```python
#  check outlier using One Variable
plt.figure()
sns.boxplot(data=cars, x="price")
plt.show()
```

```python
# check outlier (can have only one or Two not more than two) using two Variables 
# one  is numeric and  otherone is categorial
plt.figure(figsize=(8,8))
sns.boxplot(data=cars, x="price",y="make")
plt.show()
```
