```python
area = np.array([700,800,1000,1500,2000])
price= np.array([40,45,40,60,70])

plt.figure()
plt.plot(area,price)

plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price (Lakh)")

plt.show()
```

```python
# xlim and ylim
```

```python
plt.figure()
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price (Lakh)")

plt.xlim(400,2500)
plt.ylim(30,80)

plt.show()
```

```python
# xticks and yticks
```

```python
plt.figure()
plt.plot(area,price)
plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price (Lakh)")

plt.xticks(rotation=45)
plt.yticks(rotation=90)

plt.show()
```

```python
# Multiple plots
```

```python
area = np.array([700,800,1000,1500,2000])
price_mumbai= np.array([40,45,40,60,70])
price_thane= np.array([35,41,42,58,65])

plt.figure()

plt.plot(area,price_mumbai,color="yellow")
plt.plot(area,price_thane,color="red")

plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price (Lakh)")

plt.legend(labels=["Mumbai","Thane"])

plt.show()
```

## Histogram
### <b> Frequency plot for continous data

```python
price = np.array([35,41,45,47,50,51,51,59,59,60,60,61,65,70,75,76,77,78,79,80])
```

```python
# bins = 4
# (80 - 35)/4 = 11.25

#                     frequency
# bin1 -> 35 - 46  -> 3
# bin2 -> 46 - 58  -> 4
# bin3 -> 58 - 69  -> 6
# bin4 -> 69 - 80  -> 7
```

```python
# bins = 4
plt.hist(price,bins=4)
plt.show()
```

```python
# bins = 3
# (80 - 35)/3 = 15

# bin1 -> 35 - 50 -> 4
# bin2 -> 50 - 65 -> 8
# bin3 -> 65 - 80 -> 8
```

```python
plt.hist(price,bins=3)
plt.show()
# Note : by default it have 10 bins
```

```python
cars = pd.read_csv("cars.csv")
cars.head()                  
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(cars["price"])     #  by default bins=10


plt.show()
```

```python
# xavline 
```

```python
price_mean = cars["price"].mean()
price_median = cars["price"].median()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(cars["price"])     #  by default bins=10

plt.axvline(price_mean,color="green")
plt.axvline(price_median,color="red")

plt.show()
```

```python
plt.figure()

plt.title("Width distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.hist(cars["width"])     


plt.show()
```

```python
# Pandas -> numpy -> matplotlib
```

```python
cars["width"].hist()
plt.show()
```

## Barplot
<b> Frequency plot for categorical data

```python
# dummy data
place = ["Mumbai","Thane","Kalyan","Panvel"]
no_of_flats = [800,750,900,1000]
```

```python
plt.figure()
plt.bar(place,no_of_flats)
#plt.barh(place,no_of_flats)   # horizontal

plt.title("Place vs Flat Freq")
plt.xlabel("Place")
plt.ylabel("No of Flats")
plt.show()
```

## barplot using pandas

```python
#cars["body-style"].value_counts().plot(kind="bar")
cars["body-style"].value_counts().plot.bar()
plt.show()
```

## Pie Chart
### * Frequency plot for categorical plot.
### * Frequency represented in the form of percentage

```python
plt.figure(figsize=(6,5))
cars['engine-type'].value_counts().plot(kind='pie')
plt.show()
```

```python
# autopct='%1.1f%%'
```

```python
plt.figure(figsize=(8,8))
cars['engine-type'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
# explode
```

```python
plt.figure(figsize=(6,5))
cars['engine-type'].value_counts().plot(kind='pie',explode=(0,0,0,0.4,0,0,0))
plt.show()
```

## Subplots
### * Under single figure having multiple plots

```python
area = np.array([700,800,1000,1500,2000])
price= np.array([40,45,40,60,70])
```

```python
# fig - represent whole figure
# axes - represent one plot
fig,axes = plt.subplots(2,2,figsize=(8,4))
axes[0,0].hist(cars["width"])
axes[0,1].bar(place,no_of_flats)
axes[1,0].scatter(cars["engine-size"],cars["price"])
axes[1,1].plot(area,price)
plt.show()
```
