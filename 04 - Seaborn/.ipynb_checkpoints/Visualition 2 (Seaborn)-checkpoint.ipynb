# Visualization 2 (Seaborn)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
```

## 1) Line Plot

```python
x = np.arange(10)
y = np.sqrt(x)

plt.figure(figsize=(6,6))
#plt.plot(x,y)
sns.lineplot(x,y)
plt.show()
```

## 2) Histogram (distplot)

```python
plt.figure(figsize=(6,6))
sns.distplot(df["price"])
plt.show()
```

```python
mean_width = df["width"].mean()
median_width = df["width"].median()

plt.figure(figsize=(6,6))
sns.distplot(df["width"])
plt.axvline(mean_width, color= "red")
plt.axvline(median_width, color= "green")
plt.show()
```

## 3) Countplot

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=df, y="body-style")
plt.show()
```

```python
# Adding hue
plt.figure(figsize=(6,6))
sns.countplot(data=df, x="body-style", hue = "drive-wheels")
plt.show()
```

## 4) Heatmap

```python
corr = df.corr()
```

```python
# annot - label
plt.figure(figsize=(8,8))
sns.heatmap(corr, annot = True)
plt.show()
```

```python
# cmap - color change
plt.figure(figsize=(8,8))
sns.heatmap(corr, annot = True,cmap="viridis")
plt.show()
```

## 5) Scatter Plot

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x="price", y = "engine-size")
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x="price", y = "engine-size", hue="body-style")
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=df, x="price", y = "engine-size", hue="make")
plt.show()
```

## 6) Boxplot

```python
#  check outlier using One  numeric Variable
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="price")
plt.show()
```

```python
# check outlier (can have only one or Two not more than two) using two Variables 
# one  is numeric and  otherone is categorial
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="price", y="body-style")
plt.show()
```

```python
# Subplot- matplotlib
x = [1,2,3,4]
y = [3,4,5,6]

fig,axes = plt.subplots(2,3,figsize=(8,8))

axes [0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes [0,2].bar(x,y)
axes[1,0].plot(x,y)
axes [1,1].scatter(x,y)
axes[1,2].bar(x,y)
plt.show()
```

```python
# Subplot- matplotlib and seaborn
fig,axes = plt.subplots(2,2, figsize=(20,10))

sns.distplot(df["price"], ax=axes[0,0])
sns.countplot(data=df, y="make", ax= axes[0,1])
sns.scatterplot(data=df, x="price", y="make", hue="body-style", ax=axes[1,0])
sns.boxplot(data=df, x="price", ax=axes[1,1])
plt.show()
```
