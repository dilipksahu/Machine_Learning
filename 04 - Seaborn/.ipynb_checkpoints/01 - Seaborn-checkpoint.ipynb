```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv("cars.csv")
```

```python
cars.head(2)
```

## Line Plot

```python
x = [1,2,3,4,5]
y = [11,23,37,49,51]
```

```python
plt.figure()
#plt.plot(x,y)          # Matplotlib          
sns.lineplot(x,y)       # Seaborn
plt.show()
```

## Histogram

```python
plt.figure()
#plt.hist(cars["width"])
sns.distplot(cars["width"])
plt.show()
```

```python
plt.figure()
#plt.hist(cars["price"])
sns.distplot(cars["price"],bins=15)
plt.show()
```

## Countplot (Barplot)

```python
plt.figure()
#cars["body-style"].value_counts().plot(kind="bar")
sns.countplot(data=cars, x = "body-style")
plt.show()
```

```python
plt.figure()
sns.countplot(data=cars, y = "body-style")
plt.show()
```

```python
# hue
plt.figure()
sns.countplot(data=cars, x = "body-style",hue="fuel-type")
plt.show()
```

```python
plt.figure()
sns.countplot(data=cars, x = "body-style",hue="engine-location")
plt.show()
```

```python
## There is no Pie chart in seaborn as directly 
```

## Heatmap

```python
plt.figure()
sns.heatmap(cars.corr())
plt.show()
```

```python
# annot
plt.figure()
sns.heatmap(cars.corr(),annot=True)
plt.show()
```

```python
# color plaette
plt.figure()
sns.heatmap(cars.corr(),annot=True,cmap="viridis")
plt.show()
```

## Scatter Plot

```python
plt.figure()
sns.scatterplot(data=cars,x="engine-size",y="price")
plt.show()
```

```python
# hue
plt.figure()
sns.scatterplot(data=cars,x="engine-size",y="price",hue="body-style")
plt.show()
```

## Box plot
## * To plot outlier

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=cars, x="price")
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.boxplot(data=cars, y="price")
plt.show()
```

```python
# two variable
# better to go with two variable for finding outlier
plt.figure(figsize=(8,8))
sns.boxplot(data=cars, x="price", y="make")
plt.show()
```

## Subplots

```python
fig,axes = plt.subplots(2,2,figsize=(12,8))
sns.countplot(data=cars,x="body-style",
              hue="fuel-type",ax =axes[0,0])
sns.scatterplot(data=cars,x="engine-size",y="price",hue="body-style",ax=axes[0,1])
sns.boxplot(data=cars, x="price",y="body-style", ax=axes[1,0])
sns.heatmap(cars.corr(),annot=True, ax=axes[1,1])
plt.show()
```
