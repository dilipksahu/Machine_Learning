```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
from scipy.stats import skew
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("BostonHousing.csv")
```

```python
df.columns
```

```python
def lr_model(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
    model = LinearRegression()
    model.fit(X_train,y_train)
    
    print("Modeling with ",X.columns[0])
    print("intercept: ",model.intercept_)
    print("Coef: ",model.coef_)

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)

    r2 = r2_score(y_test,y_pred)

    print("mse: {},\nrmse: {},\nr2: {}".format(mse,rmse,r2))

    # Plot the model
    plt.figure()
    plt.scatter(X_test,y_test)
    plt.plot(X_test,y_pred)
    plt.show()
    print("---------------------------------------------------------------------")
```

### Builiding each separate model

```python
for col in df.iloc[:,:-1]:
    lr_model(df[[col]],df["medv"])    
```

### Building combine model

```python
X = df.drop(["medv"],axis=1)
y = df["medv"]
```

```python
def mlr_model(x,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1)
    model = LinearRegression()
    model.fit(X_train,y_train)

    print("intercept: ",model.intercept_)
    c = -1
    for col in X:
        c = c + 1
        print(f"Coef of {col}:",model.coef_[c])

    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)

    r2 = r2_score(y_test,y_pred)

    print("\nmse: {},\nrmse: {},\nr2: {}".format(mse,rmse,r2))
```

```python
mlr_model(X,y)
```
